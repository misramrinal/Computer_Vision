Question 2: Write a program to combine two images side by side and display them using OpenCV . 

```python
img=cv2.imread("R.jpg")
img2=cv2.imread("panda.jpg")
l,r,h=img.shape
a,b,c=img2.shape
```

```python
blank_image = np.zeros((max(l,a),r+b,3), np.uint8)
blank_image[:,:,:]=(255,255,255)
for i in range (0,l):
    for j in range (0,r):
        blank_image[i][j]=img[i][j]
for i in range (0,a):
    for j in range (r,r+b):
        blank_image[i][j]=img2[i][j-r]
```

```python
cv2.imshow("hi",blank_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(blank_image)
plt.show()
```

```python
cv2.imwrite("a2.jpg",blank_image)
```

Queston 3: Write an OpenCV code to load an image, extract a specific image region, and display both the original and extracted regions using Python. 

```python
img=cv2.imread("R.jpg")
blank_image = np.zeros((l,r*2,3), np.uint8)
blank_image[:,:,:]=(255,255,255)
for i in range (0,int(l/3)):
    for j in range (0,int(r/3)):
        blank_image[i+int(l/3)][j+int(r/3)]=img[i][j]
for i in range (0,l):
    for j in range (r,r*2):
        blank_image[i][j]=img[i][j-r]
```

```python
cv2.imshow("hi",blank_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(blank_image)
plt.show()
```

Question 5: Create a program that reads an image and applies a Gaussian blur to it using the filter shown in the image below. Display both the original image and the blurred image. (To apply a filter, you have to convolve the filter with the image) 

```python
img=cv2.imread("R.jpg")
```

```python
id_kernel = np.array([[1, 2, 1], 
                    [2, 4, 2], 
                    [1, 2, 1]])
id_kernel=id_kernel/16
```

```python
flt_img = cv2.filter2D(src=img, ddepth=-1, kernel=id_kernel) 
```
